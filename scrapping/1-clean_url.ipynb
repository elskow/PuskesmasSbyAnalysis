```python
import pandas as pd
```

```python
data_scrapped = pd.read_csv("url_puskesmas.csv")
data_scrapped.head()
```

```python
data_resmi = pd.read_html(
    "https://dinkes.surabaya.go.id/portalv2/upt-dinas/puskesmas/"
)[0]
data_resmi = data_resmi.drop(columns=["NO"])
data_resmi["PUSKESMAS"] = "Puskesmas " + data_resmi["PUSKESMAS"]
data_resmi.columns = data_resmi.columns.str.lower()
data_resmi.head()
```

```python
data_resmi = data_resmi.rename(columns={"puskesmas": "name"})

data = data_resmi.merge(data_scrapped, on="name", how="left")

data[data["url"].isnull()]
```

```python
data.head()
```

```python
data = data.drop(columns=["type", "address", "phone_num"])

data.head()
```

```python
data.to_csv("cleaned_puskesmas.csv", index=False)
```
